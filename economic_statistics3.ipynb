```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, YearLocator, date2num, drange
import matplotlib.ticker as ticker
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import matplotlib.colors as mcolors
```

```python
df = pd.read_csv(r"C:\Users\daisu\Documents\季節調整済指数.csv")
df = df.drop(df.columns[[0,1,2]], axis=1)
df = df.T
df = df.rename_axis('date').reset_index()
df = df.iloc[:,:2]
df = df.rename(columns={0:'steel'})
df['date'] = pd.to_datetime(df['date'] + "01")
df = df.drop(df.index[[i for i in range(72)]])
df = df.reset_index(drop=True)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(df['date'],df['steel'], color='black')
ax.set_xlabel('year-month')
ax.set_ylabel('Industrial Production Index(seasonal)')
#ax.xaxis.set_major_locator(YearLocator(1))
ax.xaxis.set_major_formatter(DateFormatter('%Y-%m'))
fig.savefig("季節調整済指数.png")
```

```python
df = pd.read_csv(r"C:\Users\daisu\Documents\季節調整済指数.csv")
df = df.drop(df.columns[[0,2]], axis=1)
df.set_index("品目名称",inplace=True)
df = df.T
df = df.rename_axis('date').reset_index()
df['date'] = pd.to_datetime(df['date'] + "01")
df = df.drop(df.index[[i for i in range(72)]])
df = df.reset_index(drop=True)
```

```python
colors = [name for name, color in mcolors.CSS4_COLORS.items()]
by_hsv = [name for name, color in mcolors.TABLEAU_COLORS.items()]
colors.extend(by_hsv)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for i in range(1,30):
    ax.plot(df['date'],df.iloc[:,i], color=colors[i])
ax.set_xlabel('year-month')
#ax.xaxis.set_major_locator(YearLocator(1))
ax.xaxis.set_major_formatter(DateFormatter('%Y-%m'))
fig.savefig("top30.png")
```

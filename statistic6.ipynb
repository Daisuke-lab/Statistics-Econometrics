```python
df3 = pd.read_csv(r"C:\Users\daisuke\Documents\gt010106.csv")
```

```python
df3.head(5)
```

```python
fig, ax = plt.subplots(figsize=(10.0, 6.0))
t = df3['year']
y1 = df3['only_PC']
y2 = df3['only_moblie']
y3 = df3['only_TV_game']
c1,c2,c3= "blue","green","red"      # 各プロットの色
l1,l2,l3 = "only_PC","only_mobile","only_TV_and_GAME"
ax.set_xlabel('year')  # x軸ラベル
ax.set_ylabel('percentage of usage')  # y軸ラベル
ax.set_title('the change of devices')
ax.set_xticks(df3['year'])
ax.plot(t, y1, color=c1, label=l1)
ax.plot(t, y2, color=c2, label=l2)
ax.plot(t, y3, color=c3, label=l3)
ax.legend(loc=0)    # 凡例
fig.tight_layout()  # レイアウトの設定
plt.savefig('the change of devices on the internet') # 画像の保存
plt.show()
```

```python
fig = plt.figure(figsize=(20.0, 6.0))
ax = fig.add_subplot(111)
x = df3['year']
y = df3['PC_mobile']
ax.set_title("the rate of mobile and PC", fontsize = 16)
ax.set_xticks(df3['year'])
ax.plot(x, y, color = "orange")
fig.savefig("the rate of mobile and pc.png")
```

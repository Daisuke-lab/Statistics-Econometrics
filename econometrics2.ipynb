## 三年以上のパネルデータ分析（失敗）

```python
import pandas as pd
import matplotlib.pyplot as plt
from linearmodels.panel.data import PanelData
from linearmodels.panel import PanelOLS, PooledOLS, RandomEffects, compare
import statsmodels.api as sm
```

```python
df = pd.read_csv(r"http://mcobaya.web.fc2.com/tfr3_long.csv", encoding="shift-jis")
df['year'] = pd.to_datetime(df['year'], format='%Y')
df=df.set_index('year', append=True)
# #swap indexes
df.index = df.index.swaplevel(0,1)
```

```python
df_dummy = pd.get_dummies(df['pref'])  
df = pd.concat([df, df_dummy], axis=1)
# df_dummy = pd.get_dummies(df['pref'])  
# df = pd.concat([df, df_dummy2], axis=1)
```

```python
df_1980 = df[df['year'] == 1980]
df_2000 = df[df['year'] == 2000]
df_2010 = df[df['year'] == 2010]
```

```python
exog_vars = ["tfr", '東京','北海道', '青森', '岩手', '宮城', '秋田', '山形', '福島', '茨城', '栃木', '群馬', '埼玉', '千葉', '神奈川', '新潟', '富山', '石川', '福井', '山梨', '長野', '奈良', '和歌山', '鳥取', '島根', '岡山', '広島', '山口', '徳島', '香川', '愛媛', '高知', '福岡', '佐賀', '長崎', '熊本', '大分', '宮崎', '鹿児島', '沖縄']
exog = sm.add_constant(df[exog_vars])
#mod = PanelOLS(df['flp'], exog, time_effects=True)
mod  = PanelOLS(df['flp'],df[exog_vars], entity_effects=True)
fe_res = mod.fit()
print(fe_res)
```

```python
formula_fe2 = 'tfr ~ flp \
                      + EntityEffects + TimeEffects'
mod_fe = PanelOLS.from_formula(formula_fe2, data=df)
result_fe = mod_fe.fit()
print(result_fe.summary.tables[1])
```
